3. Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard. 

```python
numbers = input("Enter integers separated by space: ").split()

int_list = []
for n in numbers:
    int_list.append(int(n))

largest = int_list[0]


for num in int_list:
    if num > largest:
        largest = num

print("Largest number in the list is:", largest)
```

4. Write a program that rotates the element of a list so that the element at the first index moves to the
second index, the element in the second index moves to the third index, etc., and the element in the last
index moves to the first index. 

```python
numbers = input("Enter integers separated by space: ").split()

int_list = []
for n in numbers:
    int_list.append(int(n))

first_element = int_list[0]

for i in range(len(int_list) - 1):
    int_list[i] = int_list[i + 1]

int_list[-1] = first_element

print("List after rotation:")
print(int_list)
```

5. Write a program that input a string and ask user to delete a given word from a string. 

```python
text = input("Enter a string: ")

word_to_delete = input("Enter the word to delete: ")

words = text.split()

new_words = []
for w in words:
    if w != word_to_delete:
        new_words.append(w)

new_text = " ".join(new_words)

print("Updated string:")
print(new_text)
```

 7. Write a program with a function that accepts a string from keyboard and create a new string after
converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses."
the output should be "Stop And Smell The Roses"

```python
def capitalize_words(sentence):
    words = sentence.split()
    new_words = []

    for w in words:
        new_words.append(w[0].upper() + w[1:])

    return " ".join(new_words)


text = input("Enter a sentence: ")

result = capitalize_words(text)

print("Updated string:")
print(result)
```

8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this :
Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63 

```python
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))

matrix = []

print("Enter matrix elements row wise:")
for i in range(m):
    row = []
    for j in range(n):
        value = int(input(f"Element [{i+1}][{j+1}]: "))
        row.append(value)
    matrix.append(row)


print("\nMatrix:")
for row in matrix:
    print(*row)


for i in range(m):
    row_sum = 0
    for j in range(n):
        row_sum += matrix[i][j]
    print(f"Sum of row {i+1} = {row_sum}")
```

9. Write a program to add two matrices of size n x m.

```python
n = int(input("Enter number of rows: "))
m = int(input("Enter number of columns: "))

matrix1 = []
matrix2 = []


print("Enter elements of first matrix:")
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input(f"A[{i+1}][{j+1}]: ")))
    matrix1.append(row)


print("\nEnter elements of second matrix:")
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input(f"B[{i+1}][{j+1}]: ")))
    matrix2.append(row)


result = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(matrix1[i][j] + matrix2[i][j])
    result.append(row)


print("\nSum of matrices:")
for row in result:
    print(*row)
```

10. Write a program to multiply two matrices

```python

r1 = int(input("Enter rows of first matrix: "))
c1 = int(input("Enter columns of first matrix: "))

r2 = int(input("Enter rows of second matrix: "))
c2 = int(input("Enter columns of second matrix: "))


if c1 != r2:
    print("Matrix multiplication not possible")
else:
    matrix1 = []
    matrix2 = []

 
    print("Enter elements of first matrix:")
    for i in range(r1):
        row = []
        for j in range(c1):
            row.append(int(input(f"A[{i+1}][{j+1}]: ")))
        matrix1.append(row)

   
    print("\nEnter elements of second matrix:")
    for i in range(r2):
        row = []
        for j in range(c2):
            row.append(int(input(f"B[{i+1}][{j+1}]: ")))
        matrix2.append(row)

    
    result = []
    for i in range(r1):
        row = []
        for j in range(c2):
            row.append(0)
        result.append(row)


    for i in range(r1):
        for j in range(c2):
            for k in range(c1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]


    print("\nResultant Matrix:")
    for row in result:
        print(*row)
```
